```python
#make a dataset
transactions = [
    ['apple', 'beer', 'rice', 'meat'],
    ['apple', 'beer', 'rice'],
    ['apple', 'beer'],
    ['apple', 'mango'],
    ['beer', 'mango', 'rice'],
    ['beer', 'rice'],
    ['rice', 'meat'],
    ['mango', 'milk'],
    ['milk', 'bread'],
    ['apple', 'bread', 'milk']
]

print("Generated sample transactions:")
for t in transactions:
    print(t)
```

```python
from apyori import apriori

# Applying the Apriori algorithm
rules = apriori(transactions, min_support = 0.2, min_confidence = 0.5, min_lift = 1.2, min_length = 2)

# Converting the generator object into a list for better viewing
rules_list = list(rules)

# Displaying the first few rules
print("Number of rules found:", len(rules_list))
print("\nFirst 5 Association Rules (if available):\n")
for item in rules_list[:5]:
    # first index of the inner list
    # Contains base item and add item
    # second index of the inner list
    # Contains support, confidence and lift for the afore mentioned item

    # first occurance of the base item (antecedent)
    antecedent = tuple(item[2][0][0])
    # first occurance of the add item (consequent)
    consequent = tuple(item[2][0][1])
    # support of the itemset
    support = item[1]
    # confidence for the rule
    confidence = item[2][0][2]
    # lift for the rule
    lift = item[2][0][3]

    print("Rule: " + str(antecedent) + " -> " + str(consequent))
    print("Support: " + str(round(support, 3)))
    print("Confidence: " + str(round(confidence, 3)))
    print("Lift: " + str(round(lift, 3)))
    print("=====================================")
```

```python
import pandas as pd

# Create a list to store the extracted rule details
rules_df_list = []

# Iterate through the rules and extract relevant information
for item in rules_list:
    # Assuming each item has at least one rule (item[2] is a list of rules for the itemset)
    if item[2]: # Check if the list of rules is not empty
        for rule in item[2]:
            antecedent = ', '.join(tuple(rule[0]))
            consequent = ', '.join(tuple(rule[1]))
            support = item[1]
            confidence = rule[2]
            lift = rule[3]
            rules_df_list.append({
                'antecedents': antecedent,
                'consequents': consequent,
                'support': support,
                'confidence': confidence,
                'lift': lift
            })

# Create a Pandas DataFrame from the extracted rule details
rules_df = pd.DataFrame(rules_df_list)

print("Association Rules DataFrame:")
display(rules_df.sort_values(by='lift', ascending=False))
```
