```python
#Q6
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
print (df)
```

```python
#column name
print(df.columns)
```

```python
#Checking data type for each column
print(df.dtypes)
```

```python
# Check for NaNs in each column
nan_check = df.isna().any()

print("Columns with NaN values (True/False):")
print(nan_check)
```

```python
# 3. Replace NaN values with the most frequent item (mode) for non numerical column
#manufacturer column
most_frequent_value_manufacturer = df['Manufacturer'].mode()[0]
df['Manufacturer'].fillna(most_frequent_value_manufacturer, inplace=True)
#Model
most_frequent_value_model = df['Model'].mode()[0]
df['Model'].fillna(most_frequent_value_model, inplace=True)
#type
most_frequent_value_type = df['Type'].mode()[0]
df['Type'].fillna(most_frequent_value_type, inplace=True)
#AirBags
most_frequent_value_AirBags = df['AirBags'].mode()[0]
df['AirBags'].fillna(most_frequent_value_AirBags, inplace=True)
#DriveTrain
most_frequent_value_DriveTrain = df['DriveTrain'].mode()[0]
df['DriveTrain'].fillna(most_frequent_value_DriveTrain, inplace=True)
#Origin
most_frequent_value_Origin = df['Origin'].mode()[0]
df['Origin'].fillna(most_frequent_value_Origin, inplace=True)
#Make
most_frequent_value_Make = df['Make'].mode()[0]
df['Make'].fillna(most_frequent_value_Make, inplace=True)
#cylinders
most_frequent_value_cylinders = df['Cylinders'].mode()[0]
df['Cylinders'].fillna(most_frequent_value_cylinders, inplace=True)
#Man.trans.avail
most_frequent_value_Man_trans_avail = df['Man.trans.avail'].mode()[0]
df['Man.trans.avail'].fillna(most_frequent_value_Man_trans_avail, inplace=True)
```

```python
# Check for NaNs in each column
nan_check = df.isna().any()

print("Columns with NaN values (True/False):")
print(nan_check)
```

```python
# Filling Nan with the Mean value , for numerical column
df['Min.Price'].fillna(df['Min.Price'].mean(), inplace=True)
df['Price'].fillna(df['Price'].mean(), inplace=True)
df['Max.Price'].fillna(df['Max.Price'].mean(), inplace=True)
df['MPG.city'].fillna(df['MPG.city'].mean(), inplace=True)
df['MPG.highway'].fillna(df['MPG.highway'].mean(), inplace=True)
df['EngineSize'].fillna(df['EngineSize'].mean(), inplace=True)
df['Horsepower'].fillna(df['Horsepower'].mean(), inplace=True)
df['RPM'].fillna(df['RPM'].mean(), inplace=True)
df['Rev.per.mile'].fillna(df['Rev.per.mile'].mean(), inplace=True)
df['Fuel.tank.capacity'].fillna(df['Fuel.tank.capacity'].mean(), inplace=True)
df['Passengers'].fillna(df['Passengers'].mean(), inplace=True)
df['Length'].fillna(df['Length'].mean(), inplace=True)
df['Wheelbase'].fillna(df['Wheelbase'].mean(), inplace=True)
df['Width'].fillna(df['Width'].mean(), inplace=True)
df['Turn.circle'].fillna(df['Turn.circle'].mean(), inplace=True)
df['Rear.seat.room'].fillna(df['Rear.seat.room'].mean(), inplace=True)
df['Luggage.room'].fillna(df['Luggage.room'].mean(), inplace=True)
df['Weight'].fillna(df['Weight'].mean(), inplace=True)




```

```python
#final output after cleaning
print (df)
```
