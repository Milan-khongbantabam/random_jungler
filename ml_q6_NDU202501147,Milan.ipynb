```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('Diabetes Missing Data (1).csv')
print (df.columns)
```

```python
#replacing missing values and counting the values
df = df.replace(['?', 'missing', 'none'], np.nan)
print(df.isnull().sum())
```

```python
#repalcing Nan with the mean or mode
df['Glucose'].fillna(df['Glucose'].mean(), inplace=True)
df['Diastolic_BP'].fillna(df['Diastolic_BP'].mean(), inplace=True)
df['Skin_Fold'].fillna(df['Skin_Fold'].mean(), inplace=True)
df['Serum_Insulin'].fillna(df['Serum_Insulin'].mean(), inplace=True)
df['BMI'].fillna(df['BMI'].mean(), inplace=True)
```
